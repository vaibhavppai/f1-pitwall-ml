```python
import fastf1
import pandas as pd
from tqdm import tqdm
```

```python
rounds_by_year = {
    2022: list(range(1, 23)),
    2023: list(range(1, 23)),
    2024: list(range(1, 20)),
    2025: list(range(1, 11)) 
}
```

```python
#Get laps data for the tracks    
# for year in tqdm(rounds_by_year, desc="GET laps by RACE"):
year = 2025
for round_num in rounds_by_year[year]:
    try:
        session = fastf1.get_session(year, round_num, 'R')
        session.load(telemetry=False)

        laps = session.laps.copy()

        for col in laps.columns:
            if laps[col].dtype == 'timedelta64[ns]':
                laps[col] = laps[col].dt.total_seconds()

        filename = f"{year}_round_{round_num}_laps.csv"
        laps.to_csv(filename, index=False)
    except Exception as e:
        print(f"Error in round {round_num}: {e}") 
```

```python
weather_data = []
for year in tqdm(rounds_by_year, desc="GET weather by RACE"):
    for round_num in rounds_by_year[year]:
        try:
            session = fastf1.get_session(year, round_num, 'R')
            session.load(telemetry=False, laps=False, weather=True)

            weather = session.weather_data

            for col in weather.columns:
                if weather[col].dtype == 'timedelta64[ns]':
                    weather[col] = weather[col].dt.total_seconds()

            filename = f"{year}_round_{round_num}_weather.csv"
            weather.to_csv(filename, index=False)
        except Exception as e:
            print(f"Error in round {round_num}: {e}") 
    
    
```
